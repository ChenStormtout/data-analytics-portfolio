```python
import pandas as pd
import numpy as np


```

```python
df = pd.read_csv("Amazon Sale Report cleaned.csv", sep=",")

print("DATA AWAL:")
display(df)
```

```python
df.duplicated().sum()

df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# too much missing value


df = df.drop(columns=["fulfilled-by", "Unnamed: 22"])
```

```python
# 1. Binary: Pakai promosi atau tidak
df['has_promotion'] = df['promotion-ids'].notna().astype(int)

# 2. Tipe promosi
def classify_promotion(x):
    if pd.isna(x):
        return 'No Promotion'
    elif 'Free Shipping' in str(x):
        return 'Free Shipping'
    elif 'PLCC' in str(x):
        return 'Credit Card Financing'
    else:
        return 'Other'

df['promotion_type'] = df['promotion-ids'].apply(classify_promotion)

# 3. Jumlah promosi yang dipakai (untuk multiple promotions)
df['promotion_count'] = df['promotion-ids'].apply(
    lambda x: len(str(x).split(',')) if pd.notna(x) else 0
)

# Setelah itu baru drop column asli
df.drop('promotion-ids', axis=1, inplace=True)
```

```python
df['currency'] = df['currency'].fillna(df['currency'].mode()[0])

```

```python
df['Courier Status'] = df['Courier Status'].fillna(df['Courier Status'].mode()[0])
```

```python
# drop 

df = df.dropna(subset=['ship-city', 'ship-state', 'ship-country'])
```

```python

import re

def clean_city(x):
    if pd.isna(x):
        return x
    x = x.lower()
    x = re.sub(r'[^a-z\s]', ' ', x)
    x = re.sub(r'\s+', ' ', x)
    return x.strip()

df['ship_city_clean'] = df['ship-city'].apply(clean_city)
```

```python


cities = df['ship_city_clean'].value_counts().index.tolist()
```

```python
from rapidfuzz import fuzz, process
import pandas as pd

unique_cities = df['ship_city_clean'].dropna().unique().tolist()
clusters = {}

for city in unique_cities:
    found = False
    for key in clusters:
        if fuzz.token_sort_ratio(city, key) >= 90:  # threshold bisa diubah
            clusters[key].append(city)
            found = True
            break
    if not found:
        clusters[city] = [city]

# buat mapping dari setiap variasi ke cluster key
city_map = {var: key.upper() for key, vars in clusters.items() for var in vars}

df['ship_city_norm'] = df['ship_city_clean'].map(city_map)
```
